```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (9.5,5))
sns.histplot(df['EducationBackground'])

#The below viz shows the most common educational background is the life science, and the second most is medical background
```

```python
plt.figure(figsize = (9.5,5))
sns.barplot(data = df, x ='EmpDepartment', y = 'PerformanceRating')

#From the viz we can see that Development Department on 1st position, DS on the 2nd postion following up with HR on 3rd
```

```python
sns.countplot(data = df, x= 'Gender')
#from the viz its clear 0=Male are more in comparison to female
```

```python
sns.lineplot(x = df["DistanceFromHome"], y = df["PerformanceRating"])
#Employees far are having less performance rating
```

```python
sns.pointplot(y = df['Overall rating'], x = df['YearsSinceLastPromotion'])

#We can see from viz that rating wasnt affected as it went down a bit, and again cathed the top rating, 
#this shows promotion isnt an issue
```

```python
#NO OF EMPLOYEES BELOW AVG RATING?
X = 2.76
below_avg = df[df['Overall rating']<X]
len(below_avg['EmpNumber'])
```

```python
#3 FACTOR AFFECTING THE PERFORMANCE OF EMPLOYESS? acording to correlation coefficient
# 1st EmpEnvironmentSatisfaction - 0.347
# 2nd EmpWorkLifeBalance 0.054
# 3rd EmpEducationLevel 0.0211
below_avg.corr()
```

```python
#We can easily see there is high correlation between 'Enviroment satisfaction'  and Performance rating, with value 0.3472
# therefore EmpEnvironmentSatisfaction is the key factor for low performance

```
